```python
import pandas as pd, matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np


df=pd.read_csv('Mg_l4_tetraloop_low_K_conc/fold_2_unfold/dist_Mg-ON_biased_upto-6195ns.dat', sep='\s+', header=None).dropna()
fig, ax = plt.subplots(3,3,figsize=(20,12),dpi=800,constrained_layout=True)

var = [df.iloc[:,i] for i in np.arange(1,len(df.axes[1] ) )]
var7=[var[8*i:8*(i+1)] for i in np.arange(9)]

t=df.iloc[:,0]


l=[]

for i in np.arange(3):
    for j in np.arange(3):
        l.append(ax[i,j])
        
a=['r','b','g','k','y','c','m','maroon'] 
lb=['r','b','g','k','y','c','m','maroon']

i1=1

for i,j in zip(var7,l):
    i2=0
    for m in i:
        j.plot(t,m,label=i1, color=a[i2])
        j.set_yticks(np.arange(0.18, 0.42, 0.02))
        j.set_ylim([0.18,0.42])
        j.legend()
        i1+=1
        i2+=1

plt.savefig('Mg_l4_tetraloop_low_K_conc/fold_2_unfold/dis_plot_Mg-ON_biased_upto-6195ns.pdf', format="pdf", dpi=200, bbox_inches='tight')
plt.show()
```
